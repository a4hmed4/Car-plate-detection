```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply adaptive thresholding to binarize the image
    threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

    return threshold

def segment_horizontal(image):
    # Calculate the horizontal projection of the image
    projection = np.sum(image, axis=1)

    # Define a threshold to identify regions with a high concentration of black pixels
    threshold = 0.8 * np.max(projection)

    # Identify regions with a high concentration of black pixels
    peaks = np.where(projection > threshold)[0]

    return peaks

def segment_vertical(image, line_position):
    # Create two regions based on the black line
    region1 = image[:line_position, :]
    region2 = image[line_position:, :]

    # Calculate the vertical projection of the second region
    projection_second = np.sum(region2, axis=0)

    # Define a threshold to identify regions with a high concentration of black pixels vertically
    threshold_second = 0.8 * np.max(projection_second)

    # Identify regions with a high concentration of black pixels vertically in the second region
    peaks_second = np.where(projection_second > threshold_second)[0]

    return peaks_second, region2

def apply_contour_segmentation(image, target_size=(30, 30)):
    # Convert the image to grayscale
    char_gray = image.copy()

    # Apply thresholding to convert the image to binary
    _, binary = cv2.threshold(char_gray, 128, 255, cv2.THRESH_BINARY_INV)

    # Find contours in the binary image
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Set filtering criteria (you can adjust these values based on your image characteristics)
    min_contour_area = 50
    min_contour_aspect_ratio = 0.2
    max_contour_aspect_ratio = 5.0

    # Filter and sort contours based on x-coordinate
    contours = sorted(contours, key=lambda c: cv2.boundingRect(c)[0])

    # Create a list to store character images
    characters = []

    # Iterate through each contour and extract individual characters
    for i, contour in enumerate(contours):
        # Get the bounding box of the contour
        x, y, w, h = cv2.boundingRect(contour)

        # Filter contours based on area and aspect ratio
        contour_area = cv2.contourArea(contour)
        aspect_ratio = float(w) / h

        if min_contour_area < contour_area < float('inf') and \
           min_contour_aspect_ratio < aspect_ratio < max_contour_aspect_ratio:
            # Extract the individual character from the image
            character = char_gray[y:y + h, x:x + w]

            # Resize the character to the target size
            character = cv2.resize(character, target_size)

            # Save the character image
            characters.append(character)

    return characters

def segment_and_apply_contour(image_path):
    # Read the image
    img = cv2.imread(image_path)

    # Preprocess the image
    preprocessed_img = preprocess_image(img)

    # Calculate the horizontal projection of the image
    peaks = segment_horizontal(preprocessed_img)

    # Process identified peaks
    if len(peaks) > 1:
        # Identify the position of the black line (middle point between peaks)
        line_position = (peaks[0] + peaks[-1]) // 2

        # Segment vertically and obtain the second region
        peaks_second, region2 = segment_vertical(preprocessed_img, line_position)

        # Process identified peaks in the second region
        if len(peaks_second) > 1:
            # Identify the position of the black line in the second region (middle point between peaks)
            line_position_second = (peaks_second[0] + peaks_second[-1]) // 2

            # Create two regions based on the black line in the second region
            region2_top = region2[:, :line_position_second]
            region2_bottom = region2[:, line_position_second:]

            # Apply contour segmentation to each region
            characters_top = apply_contour_segmentation(region2_top)
            characters_bottom = apply_contour_segmentation(region2_bottom)

            # Display the original images and segmented characters using Matplotlib in one line
            fig, axes = plt.subplots(1, 2 + len(characters_top) + len(characters_bottom), figsize=(20, 4))

            # Display the original images on the right
            axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            axes[0].axis('off')

            # Display the segmented characters on the left
            for i, char_image in enumerate(characters_top):
                axes[i + 1].imshow(char_image, cmap='gray_r')
                axes[i + 1].axis('off')

            for i, char_image in enumerate(characters_bottom):
                axes[len(characters_top) + i + 1].imshow(char_image, cmap='gray_r')
                axes[len(characters_top) + i + 1].axis('off')

            plt.show()

# Example usage
image_path = r"D:\car_plate_detection\EALPR_master\Plates_dataset\0003_license_plate_1.png"
segment_and_apply_contour(image_path)

```
